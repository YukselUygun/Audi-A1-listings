```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('/content/drive/MyDrive/Veri_ogrenme/Audi_A1_listings.csv')
```

```python
df.head(3)
```

```python
df=df.drop(columns=['index','href','MileageRank','PriceRank','PPYRank','Score'])
```

```python
df.head(3)
```

```python
df.info()
```

```python
df.head(3)
```

```python
df.columns=['yil','kasa','mil','motor','ps','vites','yakit','sahip','fiyat','ppy']
```

```python
df.head(3)
```

```python
df['motor']=df['motor'].str.replace('L','')
```

```python
df['motor']=pd.to_numeric(df['motor'])
```

```python
df.head(3)
```

```python
df=pd.get_dummies(df,columns=['kasa','vites','yakit'],drop_first=True)
```

```python
df.head(3)
```

```python
y=df['fiyat']
x=df.drop('fiyat',axis=1)
```

```python
lm=LinearRegression()
model=lm.fit(x,y)
```

```python
model.predict([[2017,30000,1.6,110,1,2600,0,1]])
```

```python
model.score(x,y)
```
